```python
T = data["review"]
```

```python
#Remove all non-alphabetic characters
import re
T =T.str.replace('[^A-Za-z\s]+', '')
T = T.replace(pd.np.nan, '', regex=True)
```

```python
#Make words lower case
T = T.str.lower()
```

```python
import nltk

splitwords = [nltk.word_tokenize( str(sentence) ) for sentence in T ]
print(splitwords)
```

```python
splitwords
```

```python
# converting to base words
from nltk.stem.porter import PorterStemmer
stemmer = PorterStemmer()

num_words = []

for list in range(len(splitwords)):
    num_words.append(f'{len(splitwords[list])}')
```
